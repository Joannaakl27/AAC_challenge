```python
missing_ratio = pd.DataFrame(aac_eng_df.isna().sum(), columns=['missing_count']).sort_values(by='missing_count', ascending=False)
missing_ratio['missing_count'] = missing_ratio['missing_count']/nrows
missing_ratio
# will drop breed2 and color2
```
